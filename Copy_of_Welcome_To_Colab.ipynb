```
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import GridSearchCV
import joblib
```

# **Step 2: Load the Dataset**

```
# Load the Iris dataset
iris = load_iris()
X = iris.data  # Features (sepal length, sepal width, petal length, petal width)
y = iris.target  # Labels (species of iris)
feature_names = iris.feature_names
target_names = iris.target_names

# Convert to DataFrame for better visualization
df = pd.DataFrame(X, columns=feature_names)
df['Species'] = y
print(df.head())
```

# **Step 3: Data Preprocessing**

**3.1: Check for Missing Values**

```
# Check for missing values
print(df.isnull().sum())
```

**3.2: Visualize the Data**

```
# Pairplot to visualize relationships
sns.pairplot(df, hue='Species', markers=['o', 's', 'D'])
plt.show()

```

**3.3: Feature Scaling**

```
# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

**Step 4: Split the Dataset**

```
# Split the dataset (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

# **Step 5: Model Development**

**5.1: Decision Tree (DT)**

```
# Decision Tree Model
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)

# Predictions
y_pred_dt = dt_model.predict(X_test)

# Evaluate the model
print("Decision Tree Accuracy:", accuracy_score(y_test, y_pred_dt))
print("Classification Report:\n", classification_report(y_test, y_pred_dt, target_names=target_names))
```

**5.2: K-Nearest Neighbors (KNN)**

```
# KNN Model
knn_model = KNeighborsClassifier(n_neighbors=3)
knn_model.fit(X_train, y_train)

# Predictions
y_pred_knn = knn_model.predict(X_test)

# Evaluate the model
print("KNN Accuracy:", accuracy_score(y_test, y_pred_knn))
print("Classification Report:\n", classification_report(y_test, y_pred_knn, target_names=target_names))
```

**5.3: Logistic Regression**

```
# Logistic Regression Model
lr_model = LogisticRegression(random_state=42)
lr_model.fit(X_train, y_train)

# Predictions
y_pred_lr = lr_model.predict(X_test)

# Evaluate the model
print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_lr))
print("Classification Report:\n", classification_report(y_test, y_pred_lr, target_names=target_names))
```

**5.4: Support Vector Machine (SVM)**

```
# SVM Model
svm_model = SVC(random_state=42)
svm_model.fit(X_train, y_train)

# Predictions
y_pred_svm = svm_model.predict(X_test)

# Evaluate the model
print("SVM Accuracy:", accuracy_score(y_test, y_pred_svm))
print("Classification Report:\n", classification_report(y_test, y_pred_svm, target_names=target_names))
```

**5.5: Random Forest**

```
# Random Forest Model
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

# Predictions
y_pred_rf = rf_model.predict(X_test)

# Evaluate the model
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
print("Classification Report:\n", classification_report(y_test, y_pred_rf, target_names=target_names))
```

# **Step 6: Model Optimization (Hyperparameter Tuning)**

**6.1: Decision Tree Tuning**

```
# Define parameter grid
param_grid_dt = {
    'max_depth': [3, 5, 7, None],
    'min_samples_split': [2, 5, 10]
}

# GridSearchCV for Decision Tree
grid_dt = GridSearchCV(DecisionTreeClassifier(random_state=42), param_grid_dt, cv=5)
grid_dt.fit(X_train, y_train)

# Best parameters and accuracy
print("Best Decision Tree Parameters:", grid_dt.best_params_)
print("Best Decision Tree Accuracy:", grid_dt.best_score_)
```

**6.2: KNN Tuning**

```
# Define parameter grid
param_grid_knn = {
    'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance']
}

# GridSearchCV for KNN
grid_knn = GridSearchCV(KNeighborsClassifier(), param_grid_knn, cv=5)
grid_knn.fit(X_train, y_train)

# Best parameters and accuracy
print("Best KNN Parameters:", grid_knn.best_params_)
print("Best KNN Accuracy:", grid_knn.best_score_)
```

# **Step 7: Test the Optimized Models**

```
# Test Decision Tree
best_dt = grid_dt.best_estimator_
y_pred_dt_best = best_dt.predict(X_test)
print("Optimized Decision Tree Accuracy:", accuracy_score(y_test, y_pred_dt_best))

# Test KNN
best_knn = grid_knn.best_estimator_
y_pred_knn_best = best_knn.predict(X_test)
print("Optimized KNN Accuracy:", accuracy_score(y_test, y_pred_knn_best))
```

# **Step 8: Save the Trained Models**

```
# Save Decision Tree model
joblib.dump(best_dt, 'decision_tree_model.pkl')

# Save KNN model
joblib.dump(best_knn, 'knn_model.pkl')
```

# **Step 9: Visualize Results**

```
# Confusion Matrix for Decision Tree
cm_dt = confusion_matrix(y_test, y_pred_dt)
sns.heatmap(cm_dt, annot=True, fmt='d', cmap='Blues', xticklabels=target_names, yticklabels=target_names)
plt.title('Decision Tree Confusion Matrix')
plt.show()
```

# **9.2: Compare Model Accuracies**

```
# Compare model accuracies
models = ['Decision Tree', 'KNN', 'Logistic Regression', 'SVM', 'Random Forest']
accuracies = [accuracy_score(y_test, y_pred_dt), accuracy_score(y_test, y_pred_knn), accuracy_score(y_test, y_pred_lr), accuracy_score(y_test, y_pred_svm), accuracy_score(y_test, y_pred_rf)]

plt.bar(models, accuracies, color=['blue', 'green', 'red', 'purple', 'orange'])
plt.title('Model Accuracies')
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.ylim(0.9, 1.0)
plt.show()
```
